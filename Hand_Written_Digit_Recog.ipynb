```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras

from sklearn.model_selection import train_test_split
```

```python
train=pd.read_csv("C:\\Users\\admin\\Desktop\\GITHUB Projects\\KK\\Hand Written Digit Recogniser\\train.csv")
test=pd.read_csv("C:\\Users\\admin\\Desktop\\GITHUB Projects\\KK\\Hand Written Digit Recogniser\\test.csv")
```

```python
train
```

```python
test
```

```python
target=train.label

train=train.drop('label',axis=1)
```

```python
train=train/255
test=test/255
```

```python
X=train
y=target

X_train,X_test,Y_train,Y_test=train_test_split(X,y,train_size=0.8,random_state=1)
```

```python
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
    
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=10)
```

```python
model.evaluate(X_test,Y_test)
```

```python
pred=model.predict(test)
```

```python
pred
```

```python
pred=[np.argmax(i) for i in pred]
```

```python
pred
```

```python
final_sub=pd.Series(pred,name='Label')
```
